```python
from datasets import load_dataset
```

```python
raw_datasets = load_dataset("xsum")
```

```python
raw_datasets["train"][0]
```

```python
from transformers import AutoTokenizer
    
tokenizer = AutoTokenizer.from_pretrained("t5-small")
tokenizer_base = AutoTokenizer.from_pretrained("t5-base")
```

```python
prefix = "summarize: "
max_input_length = 1024
max_target_length = 128

def preprocess_function(examples):
    inputs = [prefix + doc for doc in examples["document"]]
    model_inputs = tokenizer(inputs, max_length=max_input_length, truncation=True)

    # Setup the tokenizer for targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(examples["summary"], max_length=max_target_length, truncation=True)

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
tokenized_datasets = raw_datasets.map(preprocess_function, batched=True)
```

```python
tokenized_datasets['train'][0]
```

```python
from transformers import AutoModelForSeq2SeqLM, DataCollatorForSeq2Seq, Seq2SeqTrainingArguments, Seq2SeqTrainer

model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")
model_base = AutoModelForSeq2SeqLM.from_pretrained("t5-base")
```

```python
model.eval()
```

```python
model_base.eval()
```

```python
def summarize(sentencee, tokenizer, model1):
  sentence_ = prefix + sentencee
  input_ids = tokenizer(sentence_, max_length=1024, padding=True, truncation=True,
                        return_tensors='pt', add_special_tokens=True).input_ids
  out = model1.generate(input_ids, max_length=1024)
  summary = tokenizer.decode(out[0], skip_special_tokens=True)
  return summary
```

```python
# check how pipeline will work
from transformers import pipeline

summarizer = pipeline("summarization", model="t5-base", tokenizer="t5-base", framework="pt")
```

```python
summarizer(check, min_length=64, max_length=128)[0]['summary_text'] # standart pipeline min max can be changed for sentence length
```

Metrix with bleu score

```python
from nltk.translate.bleu_score import sentence_bleu
```

```python
def bleu_score(references, hypotesis, tokenizer, weights=(0.2, 0.2, 0.2, 0.2)):
  return sentence_bleu(references=[tokenizer(references, max_length=128, truncation=True).input_ids], 
                        hypothesis=tokenizer(hypotesis, max_length=128, truncation=True).input_ids, weights=weights)
```

```python
bleu_summary_1 = 0
for b in range(300):

  doc=tokenized_datasets['train'][b]['document']
  summar=tokenized_datasets['train'][b]['summary']

  summarized = summarize(doc, tokenizer, model)
  bleu_summary_1 += bleu_score(summar, summarized, tokenizer)
```

```python
avg1 = bleu_summary_1 / 300 # avg for 300 sentences
avg1
```

```python
bleu_summary_2 = 0
for b in range(300):

  doc=tokenized_datasets['train'][b]['document']
  summar=tokenized_datasets['train'][b]['summary']

  summarized = summarize(doc, tokenizer_base, model_base)
  bleu_summary_2 += bleu_score(summar, summarized, tokenizer_base)
```

```python
avg2 = bleu_summary_2 / 300
avg2
```

```python
#pipeline
bleu_summary_3 = 0
for b in range(300):

  doc=tokenized_datasets['train'][b]['document']
  summar=tokenized_datasets['train'][b]['summary']

  summarized = summarizer(doc, min_length=32, max_length=64)[0]['summary_text']
  bleu_summary_3 += bleu_score(summar, summarized, tokenizer_base)
```

```python
avg3 = bleu_summary_3 / 300
avg3
```

Pretrained models from t5 (t5-small and t5-base) show good performance on sentence summarization from "xsum" dataset || as shown in example summarization is not identical for all models, pipeline shows identical summarization, but with additional sentence, it can be changed by min max length. 300 sentences summarization with t5-base pipeline took longer time than using just model with t5-base, but bleu score with min_length=32, max_length=64 is lower, than without pipeline. bleu score can be improved by additional training
